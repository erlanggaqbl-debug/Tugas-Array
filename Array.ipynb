```python

# 1. STRUKTUR DATA (Array/List)
nama_barang = ["Meja", "Kursi"]
jumlah_stok = [5, 12]

def tampilkan_menu():
    """Menampilkan pilihan menu"""
    print("\n=== MANAJEMEN STOK BARANG ===")
    print("1. Tambah Barang")
    print("2. Lihat Stok")
    print("3. Perbarui Stok")
    print("4. Cari Barang")
    print("5. Keluar")
    print("---------------------------------------")

# 2. LOOP UTAMA PROGRAM
while True:
    tampilkan_menu()
    pilihan = input("Pilih menu (1-5): ")

    # --- 1. Tambah Barang ---
    if pilihan == '1':
        print("\n--- âž• Tambah Barang ---")
        nama = input("Nama Barang: ")
        stok = int(input("Jumlah Stok Awal: "))

        nama_barang.append(nama)
        jumlah_stok.append(stok)
        print(f"âœ… '{nama}' ditambahkan.")

    # --- 2. Lihat Stok ---
    elif pilihan == '2':
        print("\n--- ðŸ“¦ Daftar Stok ---")
        if not nama_barang:
            print("Stok kosong.")
        else:
            # Mengakses kedua array menggunakan indeks 'i'
            for i in range(len(nama_barang)):
                print(f"[{i+1}] {nama_barang[i]} | Stok: {jumlah_stok[i]}")
        print("-----------------------")

    # --- 3. Perbarui Stok ---
    elif pilihan == '3':
        print("\n--- ðŸ”„ Perbarui Stok ---")
        if not nama_barang:
             print("Stok kosong.")
             continue

        # Menampilkan daftar terlebih dahulu
        for i in range(len(nama_barang)):
            print(f"[{i+1}] {nama_barang[i]} | Stok: {jumlah_stok[i]}")

        nomor_urut = int(input("Nomor urut barang yang diubah: "))
        indeks = nomor_urut - 1 # Konversi ke indeks 0-based

        if 0 <= indeks < len(nama_barang):
            stok_baru = int(input(f"Masukkan stok baru untuk {nama_barang[indeks]}: "))

            # Memperbarui nilai pada array jumlah_stok
            jumlah_stok[indeks] = stok_baru
            print(f"âœ… Stok '{nama_barang[indeks]}' diperbarui.")
        else:
            print("âŒ Nomor urut tidak valid.")

    # --- 4. Cari Barang ---
    elif pilihan == '4':
        print("\n--- ðŸ” Cari Barang ---")
        kata_kunci = input("Masukkan nama barang: ").lower()
        ditemukan = False

        for i in range(len(nama_barang)):
            if kata_kunci in nama_barang[i].lower():
                print(f"âœ… Ditemukan: {nama_barang[i]} | Stok: {jumlah_stok[i]}")
                ditemukan = True

        if not ditemukan:
            print("âŒ Barang tidak ditemukan.")

    # --- 5. Keluar ---
    elif pilihan == '5':
        print("Program selesai.")
        break

    # --- Pilihan Invalid ---
    else:
        print("Pilihan tidak valid. Silakan coba lagi.")
```
